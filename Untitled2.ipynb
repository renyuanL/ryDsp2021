```python
%matplotlib qt
from ryDsp import *
def spec2d(X):
    
    X= X.T  # we like time-axis be horizontal axis
    
    X= np.log(X)
    
    pl.imshow(X, origin='lower', cmap= 'rainbow')
    
    pl.colorbar()
    pl.xlabel('n')
    pl.ylabel('k')
    pl.title('X= spectrogram(x)')

def make_spectrogram(ys, seg_length=1024):
    
    i, j= 0, seg_length
    step= seg_length // 2 # time-overlap
    
    spectrogram= [] 
    
    while j < len(ys):      
        segment= ys[i:j] 
        
        #spec= np.fft.fft(segment) 
        spec= np.fft.rfft(segment) 
        
        spectrogram += [spec]
        
        i += step
        j += step
        
    spectrogram= np.vstack(spectrogram)    
    spectrogram= np.abs(spectrogram)    
    return spectrogram

sr, x= genChirpSignal( T=10, f0=1000, f1=4000, style='sin')
X= make_spectrogram(x, seg_length=1024)
spec2d(X)
```
