```python
π= np.pi

def genSinSignal(T=1, f= 440, A=1, ϕ=0, sr= 44100):
    ts= np.linspace(0, T, sr*T)
    θ= 2 *π  *f *ts + ϕ
    ys= A * np.sin(θ)
    return sr, ys

def genCosSignal(T=1, f= 440, A=1, ϕ=0, sr= 44100):
    ts= np.linspace(0, T, sr*T)
    θ= 2 *π  *f *ts + ϕ
    ys= A * np.cos(θ)
    return sr, ys

sr, ys= genCosSignal()
```

```python
import ryDsp
```

```python
ryDsp.plotWav(ys, sr)
```

```python
ryDsp.playWav(ys,sr)
```

```python
def genSquareSignal(T=1, f= 440, A=1, ϕ=0, sr= 44100):

    sr, ys= genSinSignal(T, f, A, ϕ, sr)
    ys= np.sign(ys)
    return sr, ys

sr, ys= genSquareSignal(f=440)
ryDsp.playWav(ys,sr)
```

```python
ryDsp.plotWav(ys[:sr//100],sr)
```

```python
def genTriangleSignal(T=1, f= 440, A=1, ϕ=0, sr= 44100):
    ts= np.linspace(0, T, sr*T)
    θ= 2 *π  *f *ts + ϕ
    cycles= θ/(2 *π)
    frac, _ = np.modf(cycles)
    ys= np.abs(frac-.5) - .5/2
    ys= ys*4*A
    
    #ys= A * np.sin(θ)
    
    return sr, ys

sr, ys= genTriangleSignal(f=440)

ryDsp.plotWav(ys[:sr//100],sr)
ryDsp.playWav(ys,sr)
```

```python
sr, ys= genSquareSignal(f=440)

ryDsp.plotWav(ys[:sr//100],sr)
ryDsp.playWav(ys,sr)
```

```python
def genSawtoothSignal(T=1, f= 440, A=1, ϕ=0, sr= 44100):
    ts= np.linspace(0, T, sr*T)
    θ= 2 *π  *f *ts + ϕ
    cycles= θ/(2 *π)
    frac, _ = np.modf(cycles)
    ys= np.abs(frac) - .5
    ys= ys*2*A
    
    #ys= A * np.sin(θ)
    
    return sr, ys

sr, ys= genSawtoothSignal(f=440)

ryDsp.plotWav(ys[:sr//100],sr)
ryDsp.playWav(ys,sr)
```
